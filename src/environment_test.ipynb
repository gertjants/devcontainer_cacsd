````python
import control.matlab as ctrl
import matplotlib.pyplot as plt

s = ctrl.tf([1, 0], [1])
# define transfer functions
Hpsi = 20/s
Hs = 1 / (0.1*s + 1)
Hol = Hpsi * Hs

# print tf
print(Hol)

# run the sisotool which uses most control.matlab functionality
ctrl.sisotool(Hol)

# convert transfer function into state-space format, tests slycot
print( ctrl.tf2ss(Hol) )
````
